### Home Task
Apply PCA to breast cancer dataset and visualize 2 principle components.

#### Спершу отримую датасет

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd


# Load the breast cancer dataset
data = load_breast_cancer()
```

#### Застосовую PCA

```python
scaler = StandardScaler()

# Центрую дані, так щоб середнє усіх значень було на початку координат(цього вимагає PCA)
X_scaled = scaler.fit_transform(data.data)

# Зменшую вимірність до 2
pca = PCA(2)

X = pca.fit_transform(X_scaled)

plt.scatter(X[:,0], X[:,1], c=data.target, edgecolors='black')
plt.title('Homework PCA(2 components) for breast cancer dataset')
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
plt.show()
```
